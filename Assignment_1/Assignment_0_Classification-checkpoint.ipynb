# 2. Introduction of data for Classification & source

For the secomd visualization, I chose Crimes in Chicago which gives an incidents of crime that occurred in the City of Chicago from 2012 to 2016.

Below are the columns that are present in the given dataset:-

1. ID - Unique identifier for the record.
2. Case Number - Unique incident number.
3. Date - Date of an incident.
4. IUCR - The Illinois Unifrom Crime Reporting code
5. Primary Type - Type of the crime
6. Arrest - If arrest happened or not.
7. District - Location of an incident.
8. X and Y co-ordinate - X,Y co-ordinate of location where incident occured
9. Year - Year in which incident happened

Apart from above columns, below columns are also present in the data:-
Description, Latitude, Longitude, Location

### Source :

https://www.kaggle.com/currie32/crimes-in-chicago

## 1. Reading the data

Here I am reading the data from CSV file with the use of pandas read_csv function.

P.S. :- Place the file path in read_csv function while running the code so that csv file will get read.

```python
### Importing the required libraries ####
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
#########
##### Reading the data from the csv file ##########
data2 = pd.read_csv("Chicago_Crimes_2012_to_2017.csv", encoding='utf-8') # Place the file path here
data2.head()
```

## 2. Preprocessing of the data

In order to do the pre-processing task, I have followed below steps:-
1. Check for the null values, if any null data is present either remove the entries. If null entries can not be removed then pre-processing the same in order to get consistent data.
2. Checked if there are any duplicate entries present in data.
3. Converting date column into pandas date format, so that it can be used while visualising the data.
4. Eliminate irrelevant columns[Unnamed: 0] so that the visualization will be easier.


```python
### Checking for thr null values in data ######
print(data2.isnull().sum())
data2.isnull().values.any()
```

```python
##### droping the rows with missing data #####
data2.dropna(inplace=True)
data2.isnull().sum()
```

```python
#### Checking if there are any duplicate entries present in the data ####
data2.duplicated().sum()
```

```python
#### Converting the Date column into pandas datetime format ####
#### Checking if there are any duplicate entries present in the data ####
data2.Date = pd.to_datetime(data2.Date, format='%m/%d/%Y %I:%M:%S %p')
data2.index = pd.DatetimeIndex(data2.Date)
data2.head()
```

```python
######## Removing columns unnamed as it is not necessary on the visualization ########
data2.drop(columns=['Unnamed: 0'])
```

## 3. Visualization of the data

```python
#### In order to see the relation between different features ####
heat_map_2=plt.subplots(figsize=(10,10))
heat_map_2= sns.heatmap(data2.corr(),annot = True, cmap="YlGnBu")
plt.show()
```

- In order to find the most common crime occured in chicago, plotting the bar graph for all different types of crimes and the number of occurances.

```python
#### Analysing the categorized distribution of crime #####

fig, ax = plt.subplots(figsize=(10,10))
data2['Primary Type'].value_counts().plot.bar()
plt.title("Crimes")
plt.show()
```

### Observation:-

* Above graph shows that the general theft count was maximum in chicago considering all the types of crimes.
* Now plotting the distribution of crimes over the time to in order to see in which year the crime rate was maximum.

```python
##### Plotting line graph to see the pattern in rhe crime over the years. #####

plt.figure(figsize=(12,6))
data2.resample('Y').size().plot()
plt.title('Number of crimes per month (2010 - 2017)')
plt.xlabel('Months')
plt.ylabel('Number of crimes')
plt.show()
```

### Observation:-

1. Above graph pattern shows that the crime rate is decreasing over the years in chicago. 
2. Crime rate was maximum in 2012 and it has linearly decreased since then.

```python
########### Distribution of crime over each month and each day ############

data2.groupby([data2.index.month_name()]).size().plot(kind='bar')
plt.ylabel('Number')
plt.xlabel('Months')
plt.title('Number of crimes by month of the year')
plt.xticks(rotation=90)
plt.show()

## Creating an days array and using np array to store the days data. ####
days = ['Monday','Tuesday','Wednesday',  'Thursday', 'Friday', 'Saturday', 'Sunday']
data2.groupby([data2.index.dayofweek]).size().plot(kind='bar')
plt.ylabel('Day of week')
plt.xticks(np.arange(7), days)
plt.ylabel('Number of crimes')
plt.title('Number of crimes by day of week')
plt.show()
```

### Observation:-

1. The graph shows that the maximum crime was happened between May to August.
2. Also the crimes happeneed were maximum on Friday.

So far we have analysed that the theft and battery contribute more to the crimes and the crime rate is decreasing over the years with maximum occurance in month of May thriugh August.


Now we can also plot top 5 crimes occured in each month to observe which type of crimes occure in between May to August and how they are responsible in overall distribution

```python
### Separating out the crime data according to the years. ###
crimes_2012 = data2.loc['2012']
#print(crimes_2012)
crimes_2013 = data2.loc['2013']
crimes_2014 = data2.loc['2014']
crimes_2015 = data2.loc['2015']
crimes_2016 = data2.loc['2016']
crimes_2017 = data2.loc['2017']
#### Now we need only top 5 crimes hence removing all other types and keeping only 5 crime categories ####
print("----------------------------")
c_type_12 = pd.DataFrame(crimes_2012[crimes_2012['Primary Type'].isin(['THEFT','BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS', 'ASSAULT'])]['Primary Type'])
#print(c_type_12.head())
c_type_13 = pd.DataFrame(crimes_2013[crimes_2013['Primary Type'].isin(['THEFT','BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS', 'ASSAULT'])]['Primary Type'])
#print(c_type_13.head())
#print("----------------------------")
c_type_14 = pd.DataFrame(crimes_2014[crimes_2014['Primary Type'].isin(['THEFT','BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS', 'ASSAULT'])]['Primary Type'])
#print(c_type_14.head())
#print("----------------------------")
c_type_15 = pd.DataFrame(crimes_2015[crimes_2015['Primary Type'].isin(['THEFT','BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS', 'ASSAULT'])]['Primary Type'])
#print(c_type_15.head())
#print("----------------------------")
c_type_16 = pd.DataFrame(crimes_2016[crimes_2016['Primary Type'].isin(['THEFT','BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS', 'ASSAULT'])]['Primary Type'])
#print(c_type_16.head())
#print("----------------------------")
c_type_17 = pd.DataFrame(crimes_2017[crimes_2017['Primary Type'].isin(['THEFT','BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS', 'ASSAULT'])]['Primary Type'])
#print(c_type_17.head())
#print("----------------------------")
#---- New data frame needs to be generated which will have only type of crime and respective year ---- ###
dframe_c12= c_type_12.groupby([pd.Grouper(freq='M'), 'Primary Type'])
dframe_c13= c_type_13.groupby([pd.Grouper(freq='M'), 'Primary Type'])
dframe_c14= c_type_14.groupby([pd.Grouper(freq='M'), 'Primary Type'])
dframe_c15= c_type_15.groupby([pd.Grouper(freq='M'), 'Primary Type'])
dframe_c16= c_type_16.groupby([pd.Grouper(freq='M'), 'Primary Type'])
#dframe_c12.head()
#dframe_c13.head()
#dframe_c14.head()
#dframe_c15.head()
#dframe_c16.head()
final_df = dframe_c12['Primary Type'].count().unstack()
#plt.figure(figsize=(10,10))
final_df.plot(marker='o')
final_df2 = dframe_c13['Primary Type'].count().unstack()
final_df2.plot(marker='o')
final_df3 = dframe_c14['Primary Type'].count().unstack()
final_df3.plot(marker='o')
final_df4 = dframe_c15['Primary Type'].count().unstack()
final_df4.plot(marker='o')
final_df5 = dframe_c16['Primary Type'].count().unstack()
final_df5.plot(marker='o')
plt.show()
```

### Observation :-

- Above graph shows the distribution of top 5 crimes over all months for year 2012-2016.
- From march to may the battery and thief crimes have very less boundary in between them. It means that battery crime happens more in between march to may.
- In 2012, thief crimes were maximum in June/July; but in 2013 thief crimes were maximum in July/August. Similarly it was maximum in month of august in 2014. But in 2015 and 2016 more number of crimes happened in August. In short from 2012 to 2016, maximum occurence of thief crime changes from June to August.

    Now analysing the number of arrests happened in these 4 years.

```python
######### Arrest Ratio #####
print("Number of Arrests :- ")
print("-----------------------------")
print(data2['Arrest'].value_counts())
plt.figure(figsize=(7,7))
plt.pie(data2['Arrest'].value_counts(),autopct='%1.2f%%')
plt.title('Arrests')
```

### Observation :-

- From the above pie chart and the printed number of false(not arrested) and true(arrested) it can be inferred that only 26.16% percent of criminals have been arrested from 2012-2016.
- Plotting line graph for all 4 years will give more information about in which year more criminals were arrested.

```python
data2[data2.Arrest == True]['Year'].value_counts().plot.line()
plt.title("Arrests per year")
plt.show()
```

### Observation :-
- Number of arrests decreases over the time from 2012 to 2016. This might be due to the fact that the number of crimes also decreases w.r.t to time.
